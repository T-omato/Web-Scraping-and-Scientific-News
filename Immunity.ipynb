```python
import  time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup, SoupStrainer
from bs4.diagnose import diagnose
from parsel import Selector
from selenium.webdriver.common.action_chains import ActionChains

driver = webdriver.Chrome()
immunity = driver.get("https://www.cell.com/immunity/archive#decade=loi_decade_201")
immunity_content = driver.page_source
list_of_issues = SoupStrainer(class_='loi')
browse_soup = BeautifulSoup(immunity_content, 'html.parser', parse_only = list_of_issues)

## get's the link that opens up the issues for the volume
year_links = []
for link in browse_soup.find_all('a'):
    year_links.append(link.get('href'))
    
time.sleep(2)


volume_issue_links = []
for volume in year_links:
    immunity_vol = driver.get("https://www.cell.com" + volume)
    immunity_vol_content = driver.page_source
    volume_issues = SoupStrainer(id='issueName')
    browse_volume_issues = BeautifulSoup(immunity_vol_content, 'html.parser', parse_only = volume_issues)

    for links in browse_volume_issues.find_all('a'):
        volume_issue_links.append(links.get('href'))
    time.sleep(2)

titles = []
article_links = []
for article in volume_issue_links:
    article_page = driver.get('https://www.cell.com' + article)
    article_page_content = driver.page_source
    article_titles = SoupStrainer(class_='toc__item__title')
    browse_article_titles = BeautifulSoup(article_page_content, 'html.parser', parse_only = article_titles)
    
    for title in browse_article_titles.find_all('a'):
        titles.append(title.get_text())
    
    for title_link in browse_article_titles.find_all('a'):
        article_links.append(title_link.get('href'))

```

```python
article_links_copy = article_links.copy()

full_article_link = []
for i in article_links_copy:
    full_article_link.append("https://www.cell.com" + i)
```

```python
import json
article_dict = dict(zip(titles, full_article_link))
with open('Immunity.json', 'w+') as fp:
    json.dump(article_dict, fp,indent = "\n")
```
