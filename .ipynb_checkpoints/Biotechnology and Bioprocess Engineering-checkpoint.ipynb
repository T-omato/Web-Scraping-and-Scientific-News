```python
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup, SoupStrainer
from bs4.diagnose import diagnose

from selenium.webdriver.common.action_chains import ActionChains

driver = webdriver.Chrome()
developmental = driver.get("https://link.springer.com/journal/12257/volumes-and-issues")
developmental_content = driver.page_source
only_imp = SoupStrainer(class_='c-list-group__item')
volume_soup = BeautifulSoup(developmental_content, 'html.parser', parse_only = only_imp)

issue_href = []
issue_page = []

issue_link = []
issue_title = []

num_of_pages = 0
issue_url = ''
article_link = {}

def get_title_link(soup):
    '''Get\'s the article links and the article\'s title '''
    for link in soup.find_all('a'):
        issue_link.append(link.get('href'))

    for title in soup.find_all('a'):
        issue_title.append(title.get_text())


def get_num_of_pages(soup):
    '''If the issue has more than one page, then update num_of_pages'''
    global num_of_pages
    if soup.find('span'):
        num_of_pages = int(soup.span.string)
    return(num_of_pages)

def navigate_pages():
    '''given that there is more than 1 page per issue, navigate it'''
    global driver
    start = 1
    stop = num_of_pages
    step = 1
    stop +=step
    
    for i in range(start, stop, step):
        issue_page = issue_url + '/page/' + str(i)
        issue_ = driver.get(issue_page)
        issue_page = driver.page_source
       
        article = SoupStrainer(class_='title')
        article_soup = BeautifulSoup(issue_page, 'html.parser', parse_only = article)
        
        get_title_link(article_soup)

    
def access_issue(issue):
    '''For every issue navigate it\'s page and if there is only 1 page, parse it, otherwise parse it as if there were more than one page'''
    global issue_url
    global driver
    issue_url = 'https://link.springer.com' + issue
    issue = driver.get(issue_url)
    issue_content = driver.page_source
    pagination = SoupStrainer(class_='number-of-pages')
    page_soup = BeautifulSoup(issue_content, 'html.parser', parse_only = pagination)

    get_num_of_pages(page_soup)
    #print(page_soup.prettify())
    if num_of_pages != 0:
        navigate_pages()
        
    else:
        driver = webdriver.Chrome()
        article = SoupStrainer(class_='title')
        article_soup = BeautifulSoup(issue_content, 'html.parser', parse_only = article)
        get_title_link(article_soup)

    
for link in volume_soup.find_all('a'):
    issue_href.append(link.get('href'))

for i in issue_href:
    
    access_issue(i)
    article_link = dict(zip(issue_title, issue_link))
    

```

```python
import json
with open('Biotechnology and Bioprocess Engineering.json', 'w+') as fp:
    json.dump(article_link, fp, indent = "\n")
```
