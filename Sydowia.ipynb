```python
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup, SoupStrainer
from bs4.diagnose import diagnose

from selenium.webdriver.common.action_chains import ActionChains

driver = webdriver.Chrome()
sydowia = driver.get("http://www.sydowia.at/")
sydowia_content = driver.page_source
only_volumes = SoupStrainer(class_='subscriptions')
volume_soup = BeautifulSoup(sydowia_content, 'html.parser', parse_only = only_volumes)

##First get the journal archive

journal_link = []
for link in volume_soup.find_all('a'):
    journal_link.append(link.get('href'))

#Clean the links:
journal_issue_link = []
'''Remove the ".." from the beginning of the links'''
for i in journal_link:
    journal_issue_link.append(i.replace('..', ''))
    
#Splitting the volumes because after after a volume the html changes
first_set = journal_issue_link[0:7]
second_set = journal_issue_link[6:10]
third_set = journal_issue_link[10:]


##Next acces each journal and check out the issues and acces them

values = []
article_link = []
article_title = []
title_p = []
link_p = []
tit_sec_set = []
titles = []
true_titles = []

def create_val():
    '''appends to temporary list and cleans href values from front and coverpages and appends to permanent list later'''
    global article_link

    del article_link[:4]
    for j in article_link:
        if j[:11] != 'https://www': 
            if j not in values:
                values.append(j)

    article_link = []
    
def create_title_second():
    '''Removes the first 7 "texts" and appends the rest of the texts to the article_title list'''
    global tit_sec_set
    global true_titles
    global titles
        
    del tit_sec_set[:2]
    for i in tit_sec_set:
        if i not in titles:
            titles.append(i)
    
    true_titles = titles[0::2]
    del true_titles[:1]
    
    for i in true_titles:
        if i not in article_title:
            article_title.append(i)
            
    del true_titles[:]
    del titles[:]
    del tit_sec_set[:]
    
def create_art():
    '''Removes the first 7 "texts" and appends the rest of the texts to the article_title list'''
    global title_p

    del title_p[:7]
    for i in title_p:
        if i not in article_title:
            article_title.append(i)
  
    del title_p[:]
    
def create_val1():
    '''Removes the first 4 links and appends the rest to the values list'''
    global link_p
    del link_p[:4]
    for i in link_p:
        if i[:11] != 'https://www':
            if i not in values:
                values.append(i)

    del link_p[:]
    
for issue in first_set:
    sydowia_issue = driver.get("http://www.sydowia.at" + issue)
    issue_content = driver.page_source
    articles = SoupStrainer(class_='text')
    article_soup = BeautifulSoup(issue_content, 'html.parser', parse_only = articles)

    for link in article_soup.find_all('a'):
        article_link.append(link.get('href'))
        
    create_val()

    for title in article_soup.find_all('strong'):
        article_title.append(title.get_text())


for issues in second_set: 
    
    sydowia_issues = driver.get("http://www.sydowia.at" + issues)
    issue_contents = driver.page_source
    articless = SoupStrainer(class_='text')
    article_soups = BeautifulSoup(issue_contents, 'html.parser', parse_only = articless)
    
    for title in article_soups.find_all('td'):

        title_p.append(title.get_text(strip = True))
        
    create_title_second()
    
    
    for link in article_soups.find_all('a'):
        
        link_p.append(link.get('href'))
        
    create_val1()
    
for t_issue in third_set:
    
    sydowia_third = driver.get("http://www.sydowia.at" + t_issue)
    issue_third = driver.page_source
    t_article = SoupStrainer(class_='text')
    t_article_soup = BeautifulSoup(issue_third, 'html.parser', parse_only = t_article)
    
    for title in t_article_soup.find_all('p'):

        title_p.append(title.text)
        
    create_art()
    
    
    for link in t_article_soup.find_all('a'):
        
        link_p.append(link.get('href'))
        
    create_val1()
    
```

```python
import json
article_dict = {}
article_dict = dict(zip(article_title, values))
with open('Sydowia Fungus.json', 'w+') as fp:
    json.dump(article_dict, fp,indent = "\n")
print(article_dict)
```
