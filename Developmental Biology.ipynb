```python
import re, requests, time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup, SoupStrainer
from bs4.diagnose import diagnose
from parsel import Selector
from selenium.webdriver.common.action_chains import ActionChains

driver = webdriver.Chrome()
developmental = driver.get("https://www.sciencedirect.com/journal/developmental-biology/issues")
developmental_content = driver.page_source
only_imp = SoupStrainer(class_='accordion-title js-accordion-title')
volume_soup = BeautifulSoup(developmental_content, 'html.parser', parse_only = only_imp)

for sibling in volume_soup.span.next_siblings:
    print(sibling.string)

time.sleep(2)
issue_xpath = []    
action=ActionChains(driver)
def make_issue_xpath():
    for i in range(20):
        issue_xpath.append('//*[@id="0-accordion-tab-'+str(i)+'"]')
    return issue_xpath
make_issue_xpath()        

issue_href = []

def get_issues():
    
    for i in issue_xpath:    

        action.move_to_element(driver.find_element_by_xpath(i)).click().perform()
    develop_content = driver.page_source
    only_list = SoupStrainer(class_='js-issue-list-content')
    list_soup = BeautifulSoup(develop_content, 'html.parser', parse_only = only_list)
    
    for link in list_soup.find_all('a'):
        issue_href.append(link.get('href'))

get_issues()



time.sleep(2)
driver.close()
```

```python
driver = webdriver.Chrome()
article_link = {}
keys = []
values = []
temp_val = []
temp_key = []

def create_val():
    '''appends to temporary list and cleans href values from front and coverpages and appends to permanent list later'''
    global temp_val
    #print(temp_val)
    del temp_val[0:4]
    for j in temp_val:
        if j[-4:] != '.pdf':
            if j not in values:
                values.append(j)

    temp_val = []


def create_key():
    '''appends to temporary list and cleans "titles" from frony and coverpages and appends to permanent list later'''
    global temp_key
    del temp_key[0:2]
    for i in temp_key:
        keys.append(i)
    temp_key = []
    
def access_issue(issue):
    
    issue_url = 'https://www.sciencedirect.com' + issue
    issue_ = driver.get(issue_url)
    issue_content = driver.page_source
       
    article = SoupStrainer(class_='js-article article-content')
    article_soup = BeautifulSoup(issue_content, 'html.parser', parse_only = article)

    for string in article_soup.find_all(class_='js-article-title'):
        temp_key.append(string.get_text())

    for link in article_soup.find_all('a'):
        temp_val.append(link.get('href'))
    
    create_key()
    create_val()
    
    
for i in issue_href:
    access_issue(i)
    article_link = dict(zip(keys, values))
    
```
